### 1. Vergelijken**

- Laat de gebruiker twee woorden invoeren
- Geef aan welk woord het langs is
- Neem ook het geval dat ze gelijk zijn mee
- Print het resultaat

Hint: je mag in dit geval [de functie len()](https://www.w3schools.com/python/python_strings.asp) gebruiken

```python
# 1. Formulate a solution
# vraag de gebruiker om het 1e woord in te typen dit wordt word_1
# vraag de gebruiker om het 2e woord in te typen dit wordt word_2
# bereken lengte_ver word_1 - lengte word_2
# trek de woord lengtes van elkaar af.
# als de uitkomst groter is dan 0 is print dan een zn dat "XXX is langer dan YYY." 
# als de uitkomst 0 iis print dan een zn dat "XXX en YYY zijn even lang" 
# als de uitkomst kleiner is dan 0 is print dan een zn dat "YYY is langer dan XXX." 


# 2. Implement the solution

# vraag de gebruiker om het 1e woord in te typen dit wordt word_1
word_1 = input ('Please provide the 1st word to be compared')

# vraag de gebruiker om het 2e woord in te typen dit wordt word_2
word_2 = input ('Please provide the 2nd word to be compared')

# trek de woord lengtes van elkaar af.voor de variable length
length_dif = len(word_1)-len(word_2)


# als de uitkomst groter is dan 0 is print dan een zn dat "XXX is langer dan YYY." 
if length_dif > 0:
    print(f'{word_1} is langer dan {word_2}.')
# als de uitkomst 0 iis print dan een zn dat "XXX en YYY zijn even lang" 
elif length_dif == 0:
    print (f'{word_1} is even lang als {word_2}.')
# als de uitkomst kleiner is dan 0 is print dan een zn dat "YYY is langer dan XXX." 
else:
    print (f'{word_2} is langer dan {word_1}.')
```

### 3. Print een specifieke som**

- Print de som van een getal en het vorige getal
- Doe dit van nul tot aan een getal dat de gebruiker invoert

```python
# 1. Formulate a solution
# vraag de variable add_how_often
# maak een counter die bij 0 begint
# maak een tweede counter
# maak een while loop die door gaat tot add_how_often
# tel 1 bij de counter tel dat op bij de tweede counter
# print de tweede counter als de counter op add_how_often staat

# # 2. Implement the solution
# vraag de variable add_how_often
add_amnt = int(input('how often dow you want to do this addition?'))
# maak een counter die bij 0 begint
counter = 0
# maak een tweede counter
result = 0
# maak een while loop die door gaat tot add_amnt
while counter < add_amnt:
# tel 1 bij de counter tel dat op bij de tweede counter
    counter += 1
    result += counter
# print de tweede counter als de counter op add_amnt staat
print (result)
```
