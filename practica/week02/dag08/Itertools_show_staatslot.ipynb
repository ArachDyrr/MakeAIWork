```python
import random
import itertools

alf="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
cijf ="1234567890"
trekking_nr = 0

n = abs(int(input('Hoeveel trekkingen wil je zien? let op dit kan even duren. :P')))
 
staatslot= list(itertools.product(alf,alf,cijf,cijf,cijf,cijf,cijf))



for x in range(n):
    trekking_nr += 1
    trekking =random.choice(staatslot)
    staatslot.remove(trekking)
    print(f'De {trekking_nr}e trekking is : {trekking}') 
    


```
