```python
import pandas as pd
import numpy as np
```

```python
# Read csv file
# https://www.kaggle.com/datasets/elmoallistair/commodity-prices-19602021

comPr = "data/commodity_prices.csv"

 # Create Pandas DataFrames
df = pd.read_csv(comPr)

print(df)
```

Wat is de gemiddelde prijs voor ruwe olie over de jaren?

```python
# Zoek in de Pandas documentatie een geschikte functie
# Hint: https://pandas.pydata.org/docs/reference/series.html

# DataFrame.mean(axis=_NoDefault.no_default, skipna=True, level=None, numeric_only=None, **kwargs)[source]
# Return the mean of the values over the requested axis.




# Voer deze functie uit op het veld 'Crude Oil' en print het resultaat


average = df["Crude Oil"].mean()

print (f'The average price of Crude Oil was: ${average}')
```

Wanneer was de olieprijs op zijn hoogst? En op zijn laagst?

```python
# Print highest and lowest price
prod = "Crude Oil"

maxUSD=df[prod].max()
minUSD=df[prod].min()



maxi=df[prod].argmax()
mini=df[prod].argmin()

print (f'The highest {prod} price ${maxUSD} is at index {maxi} and lowest price ${minUSD} is at index {mini}')

maxyear = df["Year"][maxi]
minyear = df["Year"][mini]

print(f'Max {prod} price was in {maxyear} and the lowest price was at {minyear}.')
```

Net als SQL kan Pandas ook queries uitvoeren en filteren.

Zoek alle jaren waarin de olieprijs tussen 50 en 80 dollar was.

```python
# Hint: filter jaren waar de prijs tussen 50 en 80 was

df.query("50 < `Crude Oil` < 80")["Year"]
```

## Opdracht 3: samenstellen

Stel dat Nutella zou bestaan uit 50% cacao, 20% suiker, 5% mais en 25% kokosolie.

Verder is gegeven:

- voor deze grondstoffen is de prijs per kilogram
- een pot bevat 800 gram Nutella

Wat is dan de prijs voor een pot Nutella in elk jaar?

```python
# Voeg een kolom 'Nutella' toe aan het DataFrame:
df_copy=df.copy()


df_copy['Nutella']= np.NaN  #maakt een nieuwe colom en vult het met niets (numpy NaN)
# display (df_copy)

# Bereken de prijs voor een pot van 800 gram over de jaren heen:
# met functie van buurman. 
def func (a,b,c,d):
    return (.5*a+.2*b+.05*c+.25*d)

df_copy['Nutella2']= func(df_copy['Cocoa'],df_copy['Sugar'],df_copy['Maize'],df_copy['Coconut Oil'])
display (df_copy['Nutella2'])

# gebruikte pagina's:
# DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwargs)[source]#


# df['Nutella']=df.apply(func,axis=1)
df_copy['Nutella'] = df_copy.apply(lambda row: row['Cocoa']*.5 + row['Sugar']*.02 + row['Maize']*.05 + row ['Coconut Oil']*.25 , axis=1)



display (df_copy['Nutella'])
```

```python
df_co = df.copy()

def func (a,b,c,d):
    return (.5*a+.2*b+.05*c+.25*d)

df_co['Nutella'] = df_co.apply(lambda row: func(row['Cocoa'],row['Sugar'],row['Maize'],row ['Coconut Oil']) , axis=1)
display(df_co['Nutella'])
```

```python
df_nut = df.copy()

df_nut['Nutella']= (0.5*df_nut['Cocoa'] + 0.2*df_nut['Sugar'] + 0.05*df_nut['Maize'] + 0.25*df_nut['Coconut Oil'] )
display(df_nut['Nutella'])
```

```python
# Analyseer de curves by 'eyeballing':

# ...Op het blote oog kan ik niets zien om de eerste 9 jaar can steenkool te voorspellen. 
# ...ik vind het lastig om tabellen op het oog te vergelijken vooral daar dat m.i. makkelijker kan. Daarvoor ga ik kijken welke van de energie commodities het meest constant bewegen ten opzichte van coal door ze door elkaar te delen en die waardes in een grafiek tegenover elkaar te zetten 

# alleen crude oil en logs zijn nu zichtbaar omde jupyther file niet vol met grafieken te plempen.


df_coal['Coal/CrudeOil']=df_coal['Coal']/df_coal['Crude Oil']
# df_coal['Coal/NaturalGas']=df_coal['Coal']/df_coal['Natural Gas']
# df_coal['Coal/Gold']=df_coal['Coal']/df_coal['Gold']
# df_coal['Coal/PalmOil']=df_coal['Coal']/df_coal['Palm Oil']

# df_coal['Coal/Silver']=df_coal['Coal']/df_coal['Silver']
# df_coal['Coal/Rubber']=df_coal['Coal']/df_coal['Rubber']
df_coal['Coal/Logs']=df_coal['Coal']/df_coal['Logs']
# df_coal['Coal/CoconutOil']=df_coal['Coal']/df_coal['Coconut Oil']

df_coal.plot('Year', 'Coal/CrudeOil', kind='line')
# df_coal.plot('Year', 'Coal/NaturalGas', kind='line')
# df_coal.plot('Year', 'Coal/Gold', kind='line')
# df_coal.plot('Year', 'Coal/PalmOil', kind='line')

# df_coal.plot('Year', 'Coal/Silver', kind='line')
# df_coal.plot('Year', 'Coal/Rubber', kind='line')
df_coal.plot('Year', 'Coal/Logs', kind='line')
# df_coal.plot('Year', 'Coal/CoconutOil', kind='line')





plt.legend()
plt.show()
```

```python
# Maak een schatting van de waarden:

# ... als ik naar de tabelen van hierboven kijk lijken 'Logs' in het begin en "Crude Oil vanaf halverwege de jaren 70 de meest stabiele verhouding tot 'Coal' te hebben. ik kies ervoor om logs te gebruiken daar er zo'n enorme verschil in de verhouding van de waarde van Coal ten opzichte van crude oil zit in het begin van de jaren 70. 



# Gebruik Pandas om de data aan te vullen en op te slaan:

# ... Om de gaten in de dataframe aan te vullen ga ik de filna methode gebruiken. 
# DataFrame.fillna(value=None, *, method=None, axis=None, inplace=False, limit=None, downcast=None)
# Fill NA/NaN values using the specified method.

# DataFrame.first_valid_index()
# Return index for first non-NA value or None, if no non-NA value is found.

# df_coal['Coal']=df_coal['Coal'].fillna(df_coal)

df_coal_add = df.copy()
a= df_coal_add['Coal'].first_valid_index()
b=df_coal_add.loc[a,'Coal']
 

# def projection (b):
mod =    b/df_coal_add['Logs'][a]

df_coal_add['Coal']=df_coal_add['Coal'].fillna(mod*df_coal_add['Logs'])


display(df_coal_add['Coal'].head(12))
print(a)
print(b)



# ...
```

```python
# Gebruik Pandas om de data aan te vullen en op te slaan: 

df_coal_add.to_csv('data/commodity_prices_kung_fu_panda.csv', index=False)
```
