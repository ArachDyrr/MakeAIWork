```python
# Dataframe maken
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [
            22,
            35,
            58
        ],
        "Sex": [
            "male",
            "male",
            "female"
        ]
    }
    
)

# Standaard print
print(df)

# Vergelijk met
df   # jupyther functie
display(df)
```

```python
# Functie getRow() maken

def getRow (dataFrame, index):
    return dataFrame.iloc[index]

getRow(df, 2)
```

```python
# Een rij toevoegen
df_copy = df.copy()

# werkt niet
#df_copy.add(["Ruud",44,"Male"], axis=1)


df_copy.loc[3] = ["Ruud", 44, "male"]

display (df_copy)

df_copy.loc[0.5] = ["Hans", 53, "male"]

df_copy_copy = df_copy.sort_index().reset_index(drop=True)

display (df_copy)
display (df_copy_copy)

#
```
